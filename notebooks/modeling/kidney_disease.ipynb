<p>The aim of the dataset is to predict Chronic Kidney Disease (CKD) accurately based on some features.</p>
<p>The dataset is from <a href="https://archive.ics.uci.edu/dataset/336/chronic+kidney+disease">https://archive.ics.uci.edu/dataset/336/chronic+kidney+disease</a> and each row represent a patient&rsquo;s data. This business problem is important because contributes to the health sector on how be more precisely while trying to detect (CKD). The data has been collected for around 2 months. The formulation of this problem to a machine learning task would be a supervised learning classification task. The goal is to predict if a patient has CKD or not, so the target is the &ldquo;Class&rdquo; variable and is a dichotomic one, where (1 = CKD and 0= No CKD). At first sight, the data has some quality issues like null values within some features. Also, there are some categorical features that need to be processed for the modeling part.</p>


#### IMPORTING LIBRARIES

#### DATA CLEANING


FIRST INSIGHTS:
This 3 columns, are labeled as objects but they are numerical

- PackedCellVolume
- WhiteBloodCellCount
- RedBloodCellCount

We need to convert them to numeric (integer or float)


```python
# get null values from all columns
df.isnull().sum().sort_values(ascending = False)
```

```python
# convert the columns to numeric
df['PackedCellVolume'] = pd.to_numeric(df['PackedCellVolume'], errors='coerce')
df['WhiteBloodCellCount'] = pd.to_numeric(df['WhiteBloodCellCount'], errors='coerce')
df['RedBloodCellCount'] = pd.to_numeric(df['RedBloodCellCount'], errors='coerce')
```

```python
# separating numerical and categorical columns
categorical_cols = [col for col in df.columns if df[col].dtype == 'object']
numerical_cols = [col for col in df.columns if df[col].dtype != 'object']
```

```python
print('Categorical Columns:', categorical_cols)

print('Numerical Columns:', numerical_cols)
```

```python
# I will look for unique values in the categorical features
for i in categorical_cols:
    print(f"{i} column: {df[i].unique()}")
```

There are some columms that have not curated values, like "\tno" and "\tyes". We need to clean these columns:

- DiabetesMellitus
- CoronaryArteryDisease
- Class

After cleaning those values in that 3 columns, I will convert "Class" to a 1 and 0 values.


```python
df['DiabetesMellitus'].replace(to_replace = {'\tno':'no','\tyes':'yes',' yes':'yes'},inplace=True)

df['CoronaryArteryDisease'].replace(to_replace = '\tno', value='no', inplace= True)

df['Class'].replace(to_replace = {'ckd\t': 'ckd', 'notckd': 'not ckd'}, inplace=True)
```

```python
# Here I change the categorical values to numerical values (1 for positive and 0 for negative)
df['Class'] = df['Class'].map({'ckd': 0, 'not ckd': 1})

df['Class'] = pd.to_numeric(df['Class'], errors='coerce')
```

```python
# checking for values one more time to make sure we have curated them
for i in categorical_cols:
    print(f"{i} column: {df[i].unique()}")
```

#### PRE-PROCESSING


```python
df[numerical_cols].isnull().sum().sort_values(ascending=False)
```

```python
df[categorical_cols].isnull().sum().sort_values(ascending=False)
```

```python
# Here I define three functions to impute the missing values with the median, mean and mode
def impute_median(feature):
    median = df[feature].median()
    df[feature] = df[feature].fillna(median)

def impute_mean(feature):
    mean = df[feature].mean()
    df[feature] = df[feature].fillna(mean)

def impute_mode(feature):
    mode = df[feature].mode()[0]
    df[feature] = df[feature].fillna(mode)
```

```python
# Impute missing values with mean
for col in numerical_cols:
    impute_mean(col)
```

```python
df[numerical_cols].isnull().sum().sort_values(ascending=False)
```

```python
# Impute missing values with mode
for col in categorical_cols:
    impute_mode(col)
```

```python
df[categorical_cols].isnull().sum().sort_values(ascending=False)
```

#### FEATURE ENGINEERING


```python
# Here I see the unique values of the categorical columns for future encoding
for col in categorical_cols:
    print(f"{col} has {df[col].nunique()} va\n")
```

```python
# here I create an instance of the LabelEncoder
le = LabelEncoder()

# Here I encode the categorical columns with the LabelEncoder
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

#### DATASET SPLIT


```python
# dividing the datagrame into X colums and y column
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
# splitting data intp training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

#### MACHINE LEARNING MODELS


##### LOGISTIC REGRESSION


```python
# fitting the model
log_class = LogisticRegression(max_iter=1000, random_state=0)
log_class.fit(X_train, y_train)

# calculating the accuracy
log_class_acc = accuracy_score(y_test, log_class.predict(X_test))
# calculating the cross validation score
log_cross_val = cross_val_score(log_class, X_train, y_train, cv=5, scoring='accuracy').mean()
```

```python
# printing the results
print(f"Training Accuracy of Logistic Regression is {accuracy_score(y_train, log_class.predict(X_train))}")
print(f"Validation Accuracy of Logistic Regression is {log_cross_val}")
print(f"Test Accuracy of Logistic Regression is {log_class_acc} \n")
print(f"Confusion Matrix : \n{confusion_matrix(y_test, log_class.predict(X_test))}\n")
print(f"Classification Report :\n {classification_report(y_test, log_class.predict(X_test))}")
```

##### K-NEAREST NEIGHBORS


```python
# fitting the model
knn_class = KNeighborsClassifier()
knn_class.fit(X_train, y_train)

# calculating the accuracy
knn_class_acc = accuracy_score(y_test, knn_class.predict(X_test))
# calculating the cross validation score
knn_cross_val = cross_val_score(knn_class, X_train, y_train, cv=5, scoring='accuracy').mean()
```

```python
# printing the results
print(f"Training Accuracy of KNN is {accuracy_score(y_train, knn_class.predict(X_train))}")
print(f"Validation Accuracy of KNN is {knn_cross_val}")
print(f"Test Accuracy of KNN is {knn_class_acc} \n")

print(f"Confusion Matrix :\n{confusion_matrix(y_test, knn_class.predict(X_test))}\n")
print(f"Classification Report :\n {classification_report(y_test, knn_class.predict(X_test))}")
```

##### DECISION TREE


```python
# fitting the model
des_tree_class = DecisionTreeClassifier(random_state=0)
des_tree_class.fit(X_train, y_train)

# calculating the accuracy
des_tree_class_acc = accuracy_score(y_test, des_tree_class.predict(X_test))
# calculating the cross validation score
des_tree_cross_val = cross_val_score(des_tree_class, X_train, y_train, cv=5, scoring='accuracy').mean()
```

```python
# printing the results
print(f"Training Accuracy of Decision Tree Classifier is {accuracy_score(y_train, des_tree_class.predict(X_train))}")
print(f"Validation Accuracy of Decision Tree Classifier is {des_tree_cross_val}")
print(f"Test Accuracy of Decision Tree Classifier is {des_tree_class_acc} \n")

print(f"Confusion Matrix :\n{confusion_matrix(y_test, des_tree_class.predict(X_test))}\n")
print(f"Classification Report :\n {classification_report(y_test, des_tree_class.predict(X_test))}")
```

##### RANDOM FOREST


```python
# fitting the model
ran_for_class = RandomForestClassifier(criterion = 'entropy', max_depth = 11, min_samples_leaf = 2, min_samples_split = 3, n_estimators = 130, random_state=0)
ran_for_class.fit(X_train, y_train)

# calculating the accuracy
ran_for_class_acc = accuracy_score(y_test, ran_for_class.predict(X_test))
# calculating the cross validation score
ran_for_cross_val = cross_val_score(ran_for_class, X_train, y_train, cv=5, scoring='accuracy').mean()
```

```python
# printing the results
print(f"Training Accuracy of Random Forest Classifier is {accuracy_score(y_train, ran_for_class.predict(X_train))}")
print(f"Validation Accuracy of Random Forest Classifier is {ran_for_cross_val}")
print(f"Test Accuracy of Random Forest Classifier is {ran_for_class_acc} \n")

print(f"Confusion Matrix :\n{confusion_matrix(y_test, ran_for_class.predict(X_test))}\n")
print(f"Classification Report :\n {classification_report(y_test, ran_for_class.predict(X_test))}")
```

##### SUPPORT VECTOR CLASIFIER


```python
# fitting the model
svm_class = SVC(kernel='linear', random_state=0)
svm_class.fit(X_train, y_train)

# calculating the accuracy
svm_class_acc = accuracy_score(y_test, svm_class.predict(X_test))
# calculating the cross validation score
svm_cross_val = cross_val_score(svm_class, X_train, y_train, cv=5, scoring='accuracy').mean()
```

```python
# printing the results
print(f"Training Accuracy of SVC is {accuracy_score(y_train, svm_class.predict(X_train))}")
print(f"Validation Accuracy of SVC is {svm_cross_val}")
print(f"Test Accuracy of SVC is {svm_class_acc} \n")

print(f"Confusion Matrix :\n{confusion_matrix(y_test, svm_class.predict(X_test))}\n")
print(f"Classification Report :\n {classification_report(y_test, svm_class.predict(X_test))}")
```

##### GRID SEARCH AND HYPERPARAMETER TUNNING


##### RANDOM FOREST


```python
from sklearn.model_selection import GridSearchCV


param_grid = {
    'n_estimators': [100, 130, 150],
    'max_depth': [10, 11, 12],
    'min_samples_split': [2, 3, 5],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy'],
}

# set the classifier
ran_for_class = RandomForestClassifier(random_state=0)

# set up the grid search with 5-fold cross validation
grid_search = GridSearchCV(estimator=ran_for_class, 
                           param_grid=param_grid, 
                           cv=5, 
                           scoring='accuracy', 
                           n_jobs=-1,  # Use all processors
                           verbose=1)

# fit the grid search
grid_search.fit(X_train, y_train)
```

```python
# print the best parameters and the best accuracy score
print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation accuracy: ", grid_search.best_score_)

# use the best model to predict on the test set
best_ran_for_model = grid_search.best_estimator_
y_pred = best_ran_for_model.predict(X_test)

# calculate and print the test accuracy
test_accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy with best Random Forest: {test_accuracy}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# make a dataframe to compare the models
accuracies = pd.DataFrame({
    'Model': ['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest', 'SVM', 'Best Random Forest'],
    'Validation Accuracy': [log_cross_val, knn_cross_val, des_tree_cross_val, ran_for_cross_val, svm_cross_val, grid_search.best_score_],
    'Test Accuracy': [log_class_acc, knn_class_acc, des_tree_class_acc, ran_for_class_acc, svm_class_acc, test_accuracy]
})

accuracies
```
